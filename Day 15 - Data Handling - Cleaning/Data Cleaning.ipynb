<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> What is Data Cleaning
</h4> </div>

**Data Cleaning** (also called data cleansing or data scrubbing) is the process of detecting, correcting, and eliminating inaccuracies, inconsistencies, and errors in a dataset. 

This process involves preparing raw data for analysis by handling missing values, correcting incorrect entries, removing duplicate records, and ensuring data consistency across various fields.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Why is Data Cleaning Important for Data Analytics?
</h4> </div>

Data cleaning is crucial for data analytics because:

1. **Improved Accuracy**: Cleaning ensures that analysis is based on accurate and reliable information.
2. **Better Decision-Making**: Clean data enables relevant and valid insights, leading to better decisions.
3. **Reduction of Errors**: Dirty data can mislead, resulting in incorrect predictions or unreliable models.
4. **Efficiency**: Clean datasets make data processing faster and easier.
5. **High-Quality Models**: In machine learning, clean data results in better-performing models and more accurate predictions.



<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Methods of Data Cleaning
</h4> </div>

#### 1. Handling Missing Data
- **Deletion (Removing missing values):**
  - **Listwise Deletion**: Remove rows where any data is missing.
  - **Pairwise Deletion**: Remove specific fields with missing values, keeping the rest.
  
- **Imputation (Filling missing values):**
  - **Mean/Median/Mode Imputation**: Replace missing values with the mean, median, or mode of the column.
  - **Predictive Imputation**: Use machine learning models to predict missing values.
  - **Forward/Backward Fill**: Fill missing values by propagating the last valid observation forward or backward (used in time series).

#### 2. Correcting Inaccurate Data
- **Outlier Detection and Treatment**: Identify and address outliers based on their relevance to the dataset.
- **Removing Duplicates**: Detect and remove duplicate records to prevent overrepresentation of specific data points.
- **Data Consistency Checks** (not covered here): Ensure uniform formats across fields (e.g., consistent date formats, addresses, text cases).

#### 3. Data Transformation (not covered here)
- **Transformation**: applying mathematical operations or functions to your data to change its distribution or scale. This can include operations like logarithmic transformation, square root transformation, or more complex functions. Used to stabilize variance, make data more Gaussian-like, or make relationships more linear

- **Normalization**: Adjust values to a common scale without distorting their range (e.g., min-max scaling, z-score normalization). Used when features have different units or scales. It ensures that no single feature dominates the analysis due to its scale.
  
- **Standardization**: Convert data to a standardized format. Rescaling the values of your features to a common scale, often between 0 and 1, or to have a mean of 0 and a standard deviation of 1 (standardization)(e.g., one-hot encoding for categorical data, lowercase for text data).



#### 4. Addressing Data Inconsistencies (not covered here)
- **Format Standardization**: Ensure uniformity in formats like dates, units, or text case (e.g., lowercasing all text).
- **Dealing with Typographical Errors**: Correct any spelling or data entry mistakes.

#### 5. Handling Irrelevant Data (not covered here)
- Remove columns or rows that do not contribute to the analysis (e.g., columns with a high percentage of missing or irrelevant values).

**Using Z-Score**
- A Z-score indicates how many standard deviations a data point is from the mean of the dataset.
- Generally data points with Z-scores greater than 3 or less than -3 are considered outliers
- because, for a normally distributed dataset, about 99.7% of data points fall within 3 standard deviations of the mean (according to the empirical rule).

##### Standard Deviation
Standard deviation is a statistical measure that quantifies the amount of variation or dispersion in a set of values. A low standard deviation means that the values tend to be close to the mean (average) of the dataset, while a high standard deviation indicates that the values are spread out over a wider range. In formula terms, it's calculated as the square root of the variance, which is the average of the squared differences from the mean.

##### Empirical Rule
The empirical rule, also known as the 68-95-99.7 rule, states that for a normally distributed dataset:

- Approximately **68%** of the data falls within **1 standard deviation** of the mean.
- About **95%** of the data lies within **2 standard deviations** of the mean.
- Nearly **99.7%** of the data is within **3 standard deviations** of the mean.

This rule helps in understanding how data is spread around the mean in a normal distribution.

#### Using Cap or Winsorize Outliers
Instead of removing outliers, you can cap them at a certain threshold. This method replaces outlier values with the nearest non-outlier value (e.g., replacing values above the 95th percentile with the 95th percentile value).

````python
df
````

#### Using the Interquartile Range (IQR) method

```python
# Step 1: Calculate Q1 and Q3
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)

# Step 2: Calculate IQR
IQR = Q3 - Q1

# Step 3: Determine the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Identify outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Output the results
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
```
- The 1.5 * IQR rule is widely accepted in statistical analysis as a simple yet effective
- This threshold effectively captures values that are significantly lower than Q1 or higher than Q3
- In many real-world datasets that approximate a normal distribution, the 1.5 * IQR threshold helps identify observations that deviate significantly from the expected range of values.

### 1. Why do we use the argument ```inplace = True``` sometimes inside the ```fillna()``` method?

Inplace in general with other pandas operations, determines whether the operation modifies the data in the original DataFrame or creates a copy with the changes

Let's learn more about the fillna():
It is a pandas method used to replace missing values in dataframes or data series with a value. 

Syntax: 
```python
DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)
```
- value: The value to use for filling NaN
- method: Specifies the filling method. 
    - ```ffill``` or ```pad``` - Forward fill, propagates the last valid value forward
    - ```bfill``` or ```backfill``` - Backward fill, uses the next valid value to fill the missing value]
- inplace: Determines whether the fill is applied row-wise or column-wise
    - ```axis=0``` - (default) applies the fill along the rows (vertically)
    - ```axis=1``` - applies the fill across the columns (horizontally)
- limit: The maximum number of NaN values to fill in. limit=2, only the first two missing values will be filled
```python
df.fillna(0, limit=2)  # Only fills the first two NaN values with 0
```
- downcast: Attempts to downcast the filled data to a more appropriate data type. When you're filling missing values, the column's data type might be larger than necessary, and downcasting can help reduce memory usage by converting the data to a smaller type that can still hold all the values.
For example:
float64  float32
int64  int32 or int16
```python
df.fillna(value, downcast='infer') #convert the data to the smallest suitable type that can still hold the filled values
```

### 3. Difference bw loc and iloc

You use .loc to select rows and columns by their labels
Inclusive of both start and end labels
Syntax:
```python
df.loc[row_label, column_label]
```

Exclusive of the end index

df.iloc[row_index, column_index]

#### 4. A small program that does the following: (ADVANCED/OPTIONAL)

- check all columns
- if multiple null values
- drop row
- impute other columns with required imputations

This is how such a problem can come in an interview:

##### Problem Statement:

You are given a Pandas DataFrame `df` with missing values (`NaN`). You need to perform the following tasks:

1. **Identify rows** with more than one missing value and **drop** those rows.
2. **Impute remaining missing values** as follows:
   - **Numerical columns**: Replace missing values with the **mean** of that column.
   - **Categorical columns**: Replace missing values with the **mode** of that column.


````python
#Solution

import pandas as pd
import numpy as np

# create a sample df if not given in the question using dictionary structure
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Price': [10, np.nan, 20, 30, np.nan],
    'Quantity': [100, 50, np.nan, 200, np.nan],
    'Category': ['Electronics', np.nan, 'Home', 'Electronics', 'Furniture']
}

df = pd.DataFrame(data)
df
````

````python
# 1 Step: Drop rows with multiple NaN values

# Count NaN values per row
nan_counts = df.isna().sum(axis=1)
nan_counts
````

````python
# Drop rows where there are more than one NaN value
df_cleaned = df[nan_counts <= 1].copy()  # keep only the rows that have one or fewer missing values (NaNs)
df_cleaned
````

````python
# 2 Step: Impute missing values in remaining rows as usual

# Impute numerical columns with the mean
df_cleaned['Price'].fillna(df_cleaned['Price'].mean(), inplace=True)
df_cleaned['Quantity'].fillna(df_cleaned['Quantity'].mean(), inplace=True)
````

````python
# Impute categorical columns with the mode
df_cleaned['Category'].fillna(df_cleaned['Category'].mode()[0], inplace=True)
````

````python
# Output the final DataFrame
df_cleaned
````
