<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>  Advanced Seaborn – Overview


We’ll explore these Seaborn plots:

1. **pairplot** – Many scatterplots at once (pairwise relationships).
2. **jointplot** – Relationship between 2 variables + their distributions.
3. **boxenplot** – Like a boxplot but more detail in the tails.
4. **countplot** – Bar plot that counts categories (no manual aggregation).

Think of them as tools:

- **pairplot** → “How do all numeric columns relate to each other?”
- **jointplot** → “How do these 2 variables relate + their distributions?”
- **boxenplot** → “How do distributions differ across groups?”
- **countplot** → “How many records in each category?”.


<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  pairplot – Many Relationships at Once

**Idea:**  
`pairplot` shows:

- Scatter plots for each pair of numeric variables
- Histograms (or KDE) on the diagonal

Good for:

- Quick overview of all pairwise relationships
- Spotting correlations and clusters


```python
# 3.1 pairplot example

# We'll use a subset of numeric columns
pairplot_cols = ["total_bill", "tip", "size"]

sns.pairplot(tips[pairplot_cols])
plt.suptitle("Seaborn Pairplot – total_bill, tip, size", y=1.02)
plt.show()
```

**How to read this:**

- Diagonal: distribution of each variable
- Off-diagonal: relationship between two variables
- Look for:
  - Upward trend → positive relationship (e.g., total_bill vs tip)
  - Clusters → maybe different groups

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  jointplot – Focus on Two Variables

**Idea:**  
`jointplot` gives:

- A scatter plot in the middle
- Distributions (histograms/KDE) on top and right

Good for:

- Focusing on **one pair** of variables
- Seeing both relationship and distributions

We’ll use `total_bill` vs `tip`.

```python
# 3.2 jointplot example

sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")
plt.suptitle("Jointplot – Total Bill vs Tip", y=1.02)
plt.show()
```

You can also use other kinds, like:

- `kind="kde"` – smooth density
- `kind="hex"` – hexbin plot (good for many points)


A `jointplot` with `kind="kde"` shows **two things at once**:

1. **Middle area (2D KDE plot)**  
   - This is the colored “blob” in the center.  
   - Darker / more intense areas = **more data points** there.  
   - It shows **where most points are concentrated** for the two variables.  
   - Example with `total_bill` (x) and `tip` (y):  
     - A dark region around (20, 3) means many bills are around `$20` with `$3` tips.

2. **Top and Right curves (1D KDE plots)**  
   - Top curve: distribution of the **x-variable** alone (e.g., `total_bill`).  
   - Right curve: distribution of the **y-variable** alone (e.g., `tip`).  
   - Peaks in these curves show the **most common values**.

```python
#  jointplot with KDE

sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", fill=True)
plt.suptitle("Jointplot (KDE) – Total Bill vs Tip", y=1.02)
plt.show()
```

**How to interpret:**

- Look at the **shape** of the central blob:
  - If it stretches **bottom-left to top-right** → positive relationship.
  - If it stretches **top-left to bottom-right** → negative relationship.
  - If it is more **round / cloud-like** → weak or no clear relationship.
- Use the side curves to see:
  - Where each variable is **most common**.
  - Whether the data is **skewed** (more to left/right).


<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  boxenplot – Detailed Distribution Across Categories

**Idea:**  
`boxenplot` is like a boxplot but shows more detail about the distribution, especially in the tails.

Good for:

- Comparing distributions across categories (day, time, smoker, etc.)
- Seeing shape when you have **lots of data**

We’ll compare `total_bill` across `day`.

```python
# boxenplot example

plt.figure()
sns.boxenplot(data=tips, x="day", y="total_bill")

plt.title("Boxenplot – Total Bill by Day")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

**How to read:**

- Each day = distribution of total bills
- Taller/twisted shapes show spread and skew
- Outlying points show extreme bills


<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  countplot – Category Counts

**Idea:**  
`countplot` counts the number of rows in each category.

Good for:

- Quick frequency counts for categorical variables (day, sex, smoker, etc.)
- No need to compute `value_counts()` manually

We’ll look at number of bills per `day` and `time`.

```python
# 3.4 countplot example – counts per day

plt.figure()
sns.countplot(data=tips, x="day")

plt.title("Count of Bills by Day")
plt.xlabel("Day of Week")
plt.ylabel("Number of Bills")
plt.show()
```

```python
# 3.4 countplot with hue – day + time

plt.figure()
sns.countplot(data=tips, x="day", hue="time")

plt.title("Count of Bills by Day and Time")
plt.xlabel("Day of Week")
plt.ylabel("Number of Bills")

plt.legend(title="Time")
plt.show()
```

**Matplotlib/Seaborn**:

- Great for static images (reports, PDFs)
- But:

  - Harder to zoom, hover, filter



**Plotly**:

- Interactive charts by default:
  - Hover to see values
  - Zoom in/out
  - Turn series on/off by clicking legend
- Good for:
  - EDA
  - Presentations
  - Dashboards

We’ll use **Plotly Express (`plotly.express`, or `px`)** – a simple, high-level API.

```python
# 5. Basic Plotly Setup

import plotly.express as px

# Just reuse the 'tips' DataFrame
tips.head()
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Basic Interactive Scatter Plot

We’ll recreate: **total_bill vs tip**, colored by `day`, sized by `size`.

Concept:

- `x` = total_bill
- `y` = tip
- `color` = day
- `size` = number of people
- `hover_data` = extra columns to show on hover

```python
# 5.1 Plotly scatter – basic example

fig = px.scatter(
    data_frame=tips,
    x="total_bill",
    y="tip",
    color="day",          # color points by day

    size="size",          # bubble size = party size

    hover_data= ["sex", "time"],  # show extra info on hover

    title="Plotly Scatter: Tip vs Total Bill by Day",
)

fig.show()
```

**Try interacting with it:**

- Hover over points → see tooltips with extra info.
- Drag to select area → zoom in.
- Double-click background → reset.
- Click legend items (e.g., “Sun”) → show/hide that group.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Interactive Line Chart



The `tips` dataset is not time-series, but we can still:

- Pretend each row is an order in sequence.
- Plot **cumulative revenue** over order index.

This shows how revenue accumulates.

```python
# 5.3 Create an "order_id" and cumulative total_bill

tips_sorted = tips.sort_values(by="total_bill").reset_index(drop=True)

tips_sorted["order_id"] = tips_sorted.index + 1

tips_sorted["cum_revenue"] = tips_sorted["total_bill"].cumsum()

tips_sorted[["order_id", "total_bill", "cum_revenue"]].head()
```

```python
# 5.3 Plotly line chart – cumulative revenue

fig = px.line(
    data_frame=tips_sorted,
    x="order_id",
    y="cum_revenue",
    title="Cumulative Revenue Over Orders",
    labels = {"order_id": "Order Index", "cum_revenue": "Cumulative Revenue ($)"},
)

fig.show()
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Interactive Box Plot with Plotly

We can also do a **box plot** in Plotly to see distribution across categories.

Example:

> Distribution of `total_bill` by `day`.

```python
# 5.4 Plotly box plot – total bill by day

fig = px.box(
    data_frame=tips,

    x="day",
    y="total_bill",
    color="day",
    
    title="Distribution of Total Bill by Day",
    labels={"total_bill": "Total Bill ($)", "day": "Day of Week"},
)

fig.show()
```

**What you can infer:**

1. **Typical value (median)**  
   - Compare medians between days:
     - Higher median → that day usually has **higher bills**.

2. **Spread / variability (height of the box + whiskers)**  
   - Taller box → more variation in bills on that day.  
   - Shorter box → bills are more consistent.

3. **Outliers**  
   - Points far away → very high or very low bills.
   - These may be rare but important.

4. **Skewness**  
   - If median is closer to the bottom/top of the box or whiskers are uneven,
     the data might be **skewed**.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Faceting – Multiple Small Charts

**Faceting** = split data into small multiples based on categories.

Example:

> Tip vs total_bill, separate chart for each `time` (Lunch/Dinner).

```python
# 5.5 Facet scatter by time

fig = px.scatter(
    data_frame=tips,
    x="total_bill",
    y="tip",
    color="day",
    facet_col="time",  # separate small chart for each time
    title="Tip vs Total Bill by Time (Lunch vs Dinner)",
)

fig.show()
```

**Why do we facet?**

If we put everything in **one** scatter plot, patterns from different groups might **overlap** and hide each other.

Faceting allows us to:

- Compare patterns across groups **side by side**
- See if the relationship **changes between groups**

**1. Overall shape per facet**

- Does `tip` increase with `total_bill` in a similar way for **Lunch** and **Dinner**?
- Is one time more “dense” (more points) than the other?

**2. Range of values**

- Are **Dinner** bills usually higher than **Lunch** bills?
- Are tips larger at one time of day?

**3. Differences between groups**

- Does the **trend (slope)**, **spread**, or **density** look different?
- Example: At **Dinner**, high bills and high tips might be more common than at Lunch.

    <h3>  3D & Rotatable Visualizations with Plotly

Now let’s look at some **cool 3D / rotatable charts** in Plotly.

Why 3D?

- Sometimes we want to see **3 features at once** (x, y, z).
- 3D visuals are great for **showing clusters** and patterns.
- Plotly lets you **rotate, zoom, and pan** interactively with the mouse.

We’ll create:

1. A **3D scatter plot** using the classic `iris` dataset.  
2. A simple **3D surface plot** for a mathematical function.

```python
# Make sure Plotly is imported
import plotly.express as px
import numpy as np
import pandas as pd
```

**3D Scatter Plot – Iris Dataset**

The **iris** dataset has measurements of flowers:

- `sepal_length`
- `sepal_width`
- `petal_length`
- `species` (type of iris)

We will plot:

- x = sepal_length  
- y = sepal_width  
- z = petal_length  
- color = species  

Then you can **rotate** and explore how species form clusters.

```python
# 1. Load iris dataset (built into Plotly Express)
iris = px.data.iris()
iris.head()
```

```python
# 2. Create a 3D scatter plot

fig = px.scatter_3d(
    iris,
    x="sepal_length",
    y="sepal_width",
    z="petal_length",
    color="species",          # different colors for species
    symbol="species",         # different markers for species
    title="3D Scatter – Iris Flower Measurements",
    labels={
        "sepal_length": "Sepal Length",
        "sepal_width": "Sepal Width",
        "petal_length": "Petal Length",
        "species": "Species"
    }
)

# Show interactive figure (you can rotate this with the mouse)
fig.show()
```

**How to interact with the 3D scatter**

- **Left-click + drag** → rotate the plot around.
- **Scroll / pinch** → zoom in and out.
- **Right-click + drag** (or equivalent) → pan the view.
- Click legend items (species names) to **show/hide** specific species.

What to look for:

- Do some species form separate **clusters** in 3D?
- Are there species that overlap a lot?
- Does adding the 3rd axis (z) make separation between species **clearer** than in 2D?

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   3D Surface Plot – Simple Mathematical Hill



Now we’ll create a simple **mathematical surface**.

Think of:


**z = x^2 + y^2**


This is like a **bowl / hill** shape.

We’ll:

1. Create a grid of x and y values.
2. Compute z for each (x, y).
3. Plot it as a 3D surface you can rotate.

```python
import numpy as np
import plotly.graph_objects as go

# Create grid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Function: z = x² + y²
Z = X**2 + Y**2

# 3D surface plot
fig = go.Figure(
    data=[
        go.Surface(
            x=X,
            y=Y,
            z=Z,
            colorbar=dict(title="Z")
        )
    ]
)

fig.update_layout(
    title="3D Surface – z = x² + y²",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    )
)

fig.show()
```

```python
import plotly.graph_objects as go

# If your column is "tip", just change total_bill -> tip below
days = avg_bill_by_day["day"]
values = avg_bill_by_day["total_bill"]   # or avg_bill_by_day["tip"]

# Map days to numeric positions for the x-axis
x_positions = list(range(len(days)))
y_positions = [0] * len(days)  # all bars along y = 0

fig = go.Figure()

# Add one "bar" (vertical line) per day
for x, y, z, day in zip(x_positions, y_positions, values, days):
    fig.add_trace(go.Scatter3d(
        x=[x, x],
        y=[y, y],
        z=[0, z],
        mode="lines",
        line=dict(width=20),   # thickness of the bar
        name=day,
        showlegend=False
    ))

fig.update_layout(
    title="3D Bar-like Plot: Average Value by Day",
    scene=dict(
        xaxis=dict(
            title="Day",
            tickmode="array",
            tickvals=x_positions,
            ticktext=days
        ),
        yaxis=dict(title=""),   # you can label this if you want
        zaxis=dict(title="Average Total Bill")  # or "Average Tip"
    )
)

fig.show()
```
