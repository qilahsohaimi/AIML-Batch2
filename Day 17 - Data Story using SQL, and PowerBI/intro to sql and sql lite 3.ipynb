### Different Types of Data Storage Formats:

1. **CSV (Comma-Separated Values)**
   - **Structure:** A simple text format where each line corresponds to a row in a table, and columns are separated by commas.
   - **Usage:** Widely used for storing tabular data due to its simplicity and ease of use in spreadsheet applications (e.g., Excel).
   - **Example:**

     ```csv
     Name,Age,City
     Alice,30,New York
     Bob,25,Los Angeles
     ```



2. **JSON (JavaScript Object Notation)**
   - **Structure:** A lightweight data interchange format that uses a key-value pair structure. Data is represented as objects (enclosed in braces) and arrays (enclosed in brackets).
   - **Usage:** Commonly used for APIs and configuration files due to its readability and compatibility with many programming languages.
   - **Example:**
     ```json
     [
       {
         "Name": "Alice",
         "Age": 30,
         "City": "New York"
       },
       {
         "Name": "Bob",
         "Age": 25,
         "City": "Los Angeles"
       }
     ]
     ```


3. **XML (eXtensible Markup Language)**
   - **Structure:** A markup language that uses tags to define elements and structure data hierarchically. Each piece of data is enclosed within custom tags.
   - **Usage:** Often used in web services and for configuration files. While more verbose than JSON, it is also highly flexible and can represent complex data structures.
   - **Example:**
     ```xml
     <people>
       <person>
         <Name>Alice</Name>
         <Age>30</Age>
         <City>New York</City>
       </person>
       <person>
         <Name>Bob</Name>
         <Age>25</Age>
         <City>Los Angeles</City>
       </person>
     </people>
     ```

4. **SQL (Structured Query Language)**
   - **Structure:** A domain-specific language used for managing and manipulating relational databases. SQL scripts consist of commands that define database schema and operations on the data. Common commands include `CREATE`, `INSERT`, `SELECT`, `UPDATE`, and `DELETE`.
   - **Usage:** Used to perform various tasks, including creating database structures, inserting data, updating records, and querying data for reports. SQL is widely used across many relational database management systems (RDBMS).
   - **Example:**
     ```sql
     -- Create a new database
     CREATE DATABASE PeopleDB;

     -- Use the new database
     USE PeopleDB;

     -- Create a table named "people"
     CREATE TABLE people (
         id INT AUTO_INCREMENT PRIMARY KEY,
         name VARCHAR(100),
         age INT,
         city VARCHAR(100)
     );

     -- Insert data into the "people" table
     INSERT INTO people (name, age, city) VALUES
     ('Alice', 30, 'New York'),
     ('Bob', 25, 'Los Angeles');

     -- Query the data from the "people" table
     SELECT * FROM people;
     ```

#### 1. Introduction to SQL

##### What is SQL?
SQL (Structured Query Language) is a standardized programming language used to manage and manipulate relational databases. It allows you to perform various operations on the data, such as querying, updating, and managing databases.

### Types of SQL
SQL has several types or dialects, including:
- **MySQL:** An open-source relational database management system.
- **PostgreSQL:** An advanced, open-source object-relational database.
- **SQLite:** A lightweight, serverless SQL database commonly used for testing and prototyping.
- **Microsoft SQL Server:** A relational database management system developed by Microsoft.
- **Oracle Database:** A multi-model database management system produced by Oracle Corporation.

#### Why Use SQL?
- **Data Management:** Easily manage and manipulate large volumes of data.
- **Data Retrieval:** Efficiently retrieve specific information from databases.
- **Data Integrity:** Maintain the accuracy and consistency of data.

### Task 1: Create a Customer Database in SQLite

````python
# Import the sqlite3 library
import sqlite3

# Step 1: Connect to SQLite Database
# Create a new SQLite database called 'bookstore.db'.
# If it doesn't exist, it will be created.
connection = sqlite3.connect('bookstore.db')

# Step 2: Create a Cursor Object
# A cursor is used to execute SQL commands and fetch results.
cursor = connection.cursor()

# Step 3: Create the 'customers' Table
# Define the SQL command to create a table named 'customers'.
create_table_query = '''
CREATE TABLE IF NOT EXISTS customers (
    customer_id INTEGER PRIMARY KEY,  -- Unique identifier for each customer
    name TEXT NOT NULL,                -- Customer's name, cannot be NULL
    email TEXT NOT NULL UNIQUE,        -- Customer's email, cannot be NULL and must be unique
    phone TEXT,                        -- Customer's phone number, can be NULL
    joined_date DATE DEFAULT CURRENT_DATE  -- Date when the customer joined, defaults to the current date
);
'''

# Execute the command to create the table
cursor.execute(create_table_query)

# Step 4: Insert Sample Data into the 'customers' Table
# Define the SQL command to insert sample customer data.
insert_data_query = '''
INSERT INTO customers (customer_id, name, email, phone, joined_date) VALUES
(1, 'Alice Johnson', 'alice@example.com', '123-456-7890', '2024-01-15'),
(2, 'Bob Smith', 'bob@example.com', '234-567-8901', '2024-01-20'),
(3, 'Charlie Brown', 'charlie@example.com', '345-678-9012', '2024-02-01'),
(4, 'Diana Prince', 'diana@example.com', '456-789-0123', '2024-02-10');
'''

# Execute the command to insert the sample data
cursor.execute(insert_data_query)

# Step 5: Retrieve Customer Data
# Define the SQL command to select all records from the 'customers' table.
select_query = 'SELECT * FROM customers;'

# Execute the command and fetch all results
cursor.execute(select_query)
results = cursor.fetchall()  # Fetch all rows of the result

# Print the results to verify the data
print("Customer Data:")
for row in results:
    print(row)  # Print each customer's record

# Step 6: Update Customer Data
# Update Charlie Brown's phone number.
update_query = '''
UPDATE customers 
SET phone = '987-654-3210' 
WHERE name = 'Charlie Brown';
'''
# Execute the update command
cursor.execute(update_query)

# Verify the update by fetching all records again
cursor.execute(select_query)
updated_results = cursor.fetchall()
print("\nUpdated Customer Data:")
for row in updated_results:
    print(row)  # Print each customer's record after update

# Step 7: Delete a Customer Record
# Delete Bob Smith's record from the 'customers' table.
delete_query = '''
DELETE FROM customers 
WHERE name = 'Bob Smith';
'''
# Execute the delete command
cursor.execute(delete_query)

# Verify the deletion by fetching all records again
cursor.execute(select_query)
final_results = cursor.fetchall()
print("\nFinal Customer Data After Deletion:")
for row in final_results:
    print(row)  # Print each customer's record after deletion

# Step 8: Close the Database Connection
# Commit changes and close the connection to the database
connection.commit()  # Ensure all changes are saved
connection.close()   # Close the connection
````

#### Create a Library Database in mySQL (Optional - Advanced)

````python
# Import the required library for MySQL connection
import mysql.connector

# Establish a connection to the MySQL server
connection = mysql.connector.connect(
    host='localhost',        # The host where the MySQL server is running
    user='your_username',    # Replace with your MySQL username
    password='your_password'  # Replace with your MySQL password
)

# Create a cursor object to execute SQL commands
cursor = connection.cursor()

# Step 1: Create a new database named 'library_db'
cursor.execute("CREATE DATABASE IF NOT EXISTS library_db")
print("Database 'library_db' created successfully.")

# Step 2: Select the newly created database
cursor.execute("USE library_db")

# Step 3: Create the 'members' table
cursor.execute("""
CREATE TABLE IF NOT EXISTS members (
    member_id INT AUTO_INCREMENT PRIMARY KEY,  -- Primary key with auto-increment
    name VARCHAR(100) NOT NULL,                 -- Member's name
    email VARCHAR(100) NOT NULL UNIQUE,         -- Member's email, must be unique
    join_date DATE NOT NULL                     -- Date when the member joined
)
""")
print("Table 'members' created successfully.")

# Step 4: Create the 'books' table
cursor.execute("""
CREATE TABLE IF NOT EXISTS books (
    book_id INT AUTO_INCREMENT PRIMARY KEY,      -- Primary key with auto-increment
    title VARCHAR(255) NOT NULL,                 -- Title of the book
    author VARCHAR(100) NOT NULL,                -- Author of the book
    published_year YEAR NOT NULL,                -- Year the book was published
    available_copies INT NOT NULL                -- Number of available copies
)
""")
print("Table 'books' created successfully.")

# Step 5: Insert sample data into the 'members' table
cursor.execute("""
INSERT INTO members (name, email, join_date) VALUES
('Alice Johnson', 'alice.johnson@example.com', '2024-01-05'),
('Bob Brown', 'bob.brown@example.com', '2024-02-10'),
('Carol White', 'carol.white@example.com', '2024-03-15')
""")
print("Sample data inserted into 'members' table.")

# Step 6: Insert sample data into the 'books' table
cursor.execute("""
INSERT INTO books (title, author, published_year, available_copies) VALUES
('To Kill a Mockingbird', 'Harper Lee', 1960, 3),
('1984', 'George Orwell', 1949, 5),
('The Great Gatsby', 'F. Scott Fitzgerald', 1925, 2)
""")
print("Sample data inserted into 'books' table.")

# Commit the transaction to save changes
connection.commit()

# Step 7: Write SQL queries to retrieve information

# Query 1: List all members who joined after January 1, 2024
cursor.execute("SELECT * FROM members WHERE join_date > '2024-01-01'")
members_after_jan_2024 = cursor.fetchall()
print("\nMembers who joined after January 1, 2024:")
for member in members_after_jan_2024:
    print(member)

# Query 2: Find the total number of available copies for all books
cursor.execute("SELECT SUM(available_copies) FROM books")
total_copies = cursor.fetchone()[0]
print("\nTotal number of available copies for all books:", total_copies)

# Query 3: Get the details of the book with the highest published year
cursor.execute("SELECT * FROM books ORDER BY published_year DESC LIMIT 1")
latest_book = cursor.fetchone()
print("\nDetails of the book with the highest published year:")
print(latest_book)

# Close the cursor and connection
cursor.close()
connection.close()
````
