#### Learning about selections

````python
-- selection 
-- WHERE clause 

-- let's us keep rows that meet certain conditions

SELECT * from Employees;

SELECT name, salary, city, hiring_year FROM Employees WHERE city = 'New York';

-- get the employees with salary greater that 80K 

SELECT name, salary, city, hiring_year 
FROM Employees 
WHERE salary > 60000 AND city = 'Chicago' AND hiring_year = 2018;

-- AND OR 

-- USing IN 
SELECT name, salary, city, hiring_year 
FROM employees
WHERE city IN ('New York', 'Chicago');

-- USing between 
SELECT name, salary, city, hiring_year 
FROM employees
WHERE salary BETWEEN 60000 AND 800000
````

````python
cursor.execute('''

CREATE TABLE students(
    id INT PRIMARY KEY,
    name VARCHAR(10),
    city VARCHAR(10),
    hobby VARCHAR(10)
)

''')
````
